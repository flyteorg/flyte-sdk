# Calling Async Flyte Tasks

Jupyter supports top-level `await`, so you can call async Flyte tasks
directly. When running remotely, these are submitted to Flyte.

```python
a = 10.0
b = 20.0
```

```python
# Await async tasks — Jupyter supports top-level await
from async_example import async_double, async_add

doubled_a = await async_double(n=int(a))
doubled_b = await async_double(n=int(b))

result = await async_add(a=float(doubled_a), b=float(doubled_b))

print(f"double({a}) = {doubled_a}")
print(f"double({b}) = {doubled_b}")
print(f"add({doubled_a}, {doubled_b}) = {result}")
```

```python
from flyteplugins.papermill import record_outputs

record_outputs(result=result)
```
