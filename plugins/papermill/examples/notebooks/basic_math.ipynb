# Basic Math Notebook

A simple notebook that takes two numbers and returns their sum.
This demonstrates the basic papermill plugin workflow.

```python
# These are default values. Papermill will inject actual values at runtime.
x = 0
y = 0.0
```

```python
print(f"Computing {x} + {y}")
result = x + y
print(f"Result: {result}")
```

```python
from flyteplugins.papermill import record_outputs

record_outputs(result=float(result))
```
