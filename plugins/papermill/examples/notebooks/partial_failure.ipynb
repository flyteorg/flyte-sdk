# Partial Failure Demo

This notebook deliberately raises an error to demonstrate that the report
is rendered from the partial output — showing all cells that completed
before the failure.

Open the **Report** tab in the Flyte UI after the task fails to see the
successful cell outputs.

```python
# Default values — papermill injects actual values after this cell
n = 5
```

```python
# Step 1: some work that succeeds
print(f"Starting computation with n={n}")
partial_result = sum(range(n))
print(f"Sum of 0..{n - 1} = {partial_result}")
```

```python
# Step 2: more work that also succeeds
squares = [i**2 for i in range(n)]
print(f"Squares: {squares}")
print(f"Sum of squares: {sum(squares)}")
```

```python
# Step 3: this cell raises an error — simulating a real failure
# (e.g. bad data, network error, assertion failure)
if n < 10:
    raise ValueError(
        f"n={n} is too small for this computation. The Flyte report above shows the partial results from Steps 1 and 2."
    )

result = partial_result * 2
```

```python
from flyteplugins.papermill import record_outputs

record_outputs(result=result)
```
