# No Outputs Notebook

A notebook that performs side effects (logging, plotting, etc.) without returning outputs.
The executed notebook itself is rendered as a Flyte report.

```python
message = "hello"
count = 5
```

```python
for i in range(count):
    print(f"[{i+1}/{count}] {message}")
```

```python
print(f"Done! Printed '{message}' {count} times.")
```
