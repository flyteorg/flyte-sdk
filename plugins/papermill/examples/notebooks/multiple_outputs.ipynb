# Multiple Outputs Notebook

Demonstrates returning multiple output values from a notebook task.

```python
text = "hello world"
n = 3
```

```python
repeated = text * n
word_count = len(text.split())
char_count = len(text)

print(f"Original: {text}")
print(f"Repeated {n}x: {repeated}")
print(f"Words: {word_count}, Chars: {char_count}")
```

```python
from flyteplugins.papermill import record_outputs

record_outputs(repeated=repeated, word_count=word_count, char_count=char_count)
```
