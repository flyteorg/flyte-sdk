# Processing a Flyte File in a Notebook

Demonstrates receiving a `File` input (as a path string) and
returning both a primitive and a `File` output.

```python
# Default values — papermill injects actual values after this cell
input_file = None
```

```python
from flyte.io import File

from flyteplugins.papermill import load_file

# Reconstruct the File from the serialized path string
f = load_file(input_file)

# Read and count lines
with f.open_sync("rb") as fh:
    content = fh.read()
lines = content.decode().splitlines()
line_count = len(lines)
print(f"File has {line_count} lines")


# Create an output file with processed content
output_file = File.new_remote("processed.txt")
with output_file.open_sync("w") as fh:
    for i, line in enumerate(lines, 1):
        fh.write(f"{i}: {line}\n")

print(f"Output written to {output_file.path}")
```

```python
from flyteplugins.papermill import record_outputs

record_outputs(line_count=line_count, output_file=output_file)
```
