# Primitive Types Notebook

Demonstrates passing `bool`, `list`, and `dict` inputs to a notebook via papermill.
The notebook filters `values` using a threshold from the `options` dict,
but only when `enabled` is `True`.

```python
enabled = True
values = []
options = {}
```

```python
threshold = options.get("threshold", 0)
label = options.get("label", "unlabeled")

if enabled:
    above = [v for v in values if v > threshold]
    total = float(sum(values))
else:
    above = []
    total = 0.0

count = len(above)
print(f"label={label!r}, enabled={enabled}, total={total}, count={count}")
```

```python
from flyteplugins.papermill import record_outputs

record_outputs(count=count, total=total, label=label)
```
