# Calling Flyte Tasks from a Notebook

Demonstrates calling other Flyte tasks from within a notebook.
When running remotely, these calls are submitted to Flyte and appear
as separate nodes in the Flyte UI.

```python
a = 10
b = 20
```

```python
# Import a task defined in the same project
from tasks import double

# This runs as a local function call locally,
# and gets submitted to Flyte when running remotely.
doubled_a = double(n=a)
doubled_b = double(n=b)

print(f"double({a}) = {doubled_a}")
print(f"double({b}) = {doubled_b}")
```

```python
total = doubled_a + doubled_b
print(f"Total: {total}")
```

```python
from flyteplugins.papermill import record_outputs

record_outputs(total=total)
```
