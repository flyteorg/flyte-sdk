# Processing a DataFrame in a Notebook

Demonstrates receiving a `DataFrame` input (as a URI string) and
returning both a primitive and a `DataFrame` output.

```python
# Default values — papermill injects actual values after this cell
input_df = None
```

```python
import pandas as pd
from flyte.io import DataFrame

from flyteplugins.papermill import load_dataframe

# Reconstruct the DataFrame from the serialized URI string
df = load_dataframe(input_df)

# Materialize as pandas and inspect
pdf = await df.open(pd.DataFrame).all()
row_count = len(pdf)
print(f"DataFrame has {row_count} rows")
print(pdf)

# Filter and create output DataFrame
filtered = pdf[pdf["score"] >= 80]
print(f"Filtered to {len(filtered)} rows with score >= 80")
filtered_df = await DataFrame.from_local(filtered)
```

```python
from flyteplugins.papermill import record_outputs

record_outputs(row_count=row_count, filtered_df=filtered_df)
```
